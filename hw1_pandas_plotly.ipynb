**5. <font color='orange'>[2 ]</font>       .         **

```python
#  
data_piter['1_metr'] = data_piter['price'] / data_piter['area']
```

```python
data_piter[(data_piter['1_metr'] <= 350000) & (data_piter['1_metr'] >= 0)]["1_metr"].hist()
```

**6. <font color='orange'>[4 ]</font>           -?   ()  **

```python
#  
data_piter_real = data_piter[(data_piter['1_metr'] > 10000) & (data_piter['1_metr'] < 400000)]

data_piter_real["1_metr"] = np.round(data_piter_real["1_metr"])


fig = px.scatter_mapbox(data_piter_real.sample(500),
                        lat='geo_lat',
                        lon='geo_lon',
                        hover_name='region',
                        size='1_metr',
                        zoom=3)
fig.update_layout(mapbox_style="open-street-map")
fig.show(renderer='svg', engine='kaleido', width=900, height=650)
```
